```python
##Importerar moduler 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

##Läser in min data fil och döper den till "graph"
graph= pd.read_csv('covid_norden_3.csv')

##Gör om storleken på graphen
plt.figure(figsize=(8,5))

##lägger till titeln
plt.title('Totala corona fall uppdaterad per månad')

##skapar ett streck för varje land som är beroende av datumet, dessutom lägger till en label som
##Döper om varje streck till det landet men dessutom behöver jag denna labelnen för att plt.legend 
##ska fungera.
plt.plot(graph.date, graph.casesDenmark, label='Denmark')
plt.plot(graph.date, graph.casesFinland, label='Finland')
plt.plot(graph.date, graph.casesNorway, label='Norway')
plt.plot(graph.date, graph.casesSweden, label='Sweden')

##Här gör jag så vart sjätte tal på x axeln visas. 
plt.xticks(graph.date[::6])

##Sätter små rubriker på x och y axeln
plt.ylabel('Mått i miljoner')
plt.xlabel('Datum')

##Visar bara labelnen för varje streck
plt.legend()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

graph= pd.read_csv('covid_norden_3.csv')

plt.figure(figsize=(8,5))

plt.title('Totala döds fall uppdaterad per månad')

plt.plot(graph.date, graph.deathsDenmark, label='Denmark')
plt.plot(graph.date, graph.deathsFinland, label='Finland')
plt.plot(graph.date, graph.deathsNorway, label='Norway')
plt.plot(graph.date, graph.deathsSweden, label='Sweden')

plt.xticks(graph.date[::6])

plt.ylabel('Mått i miljoner')
plt.xlabel('Datum')

plt.legend()

plt.show()
```
